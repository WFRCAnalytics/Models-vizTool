```python
import pandas as pd
import os
import json
```

```python
# Load the JSON file
with open("../../scenarios.json", 'r') as f:
    json_data = json.load(f)

# Extract 'data' part
data = json_data['data']

# Convert to DataFrame
dfScn = pd.json_normalize(data)

# Display the DataFrame
display(dfScn)
```

```python
import pandas as pd
import os
import glob

dfRouteNames = pd.DataFrame()

# Iterate through the scenarios
for _, row in dfScn.iterrows():

    _scenarioCode = row['modVersion'] + '__' + row['scnGroup'] + '__' + str(row['scnYear'])

    _filename = '../' + _scenarioCode + "/j-transit-route-names.json"

    print(_filename)

    # Read the CSV into a dataframe
    _df = pd.read_json(_filename)

    _df.rename(columns={'Name':'RouteName','LongName':'RouteLongName'}, inplace=True)

    dfRouteNames = pd.concat([dfRouteNames, _df])

dfRouteNames = dfRouteNames.drop_duplicates()
display(dfRouteNames)

# Specify the desired file name
filename = '../../transit-route-names.json'

dfRouteNames.to_json(filename, orient='records')

print('All Done!')
```
