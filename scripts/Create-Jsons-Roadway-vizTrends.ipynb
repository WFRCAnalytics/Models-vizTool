```python
import pandas as pd
import os
import glob

# Define the subfolder path
subfolder_path = './data/segSummaries/'

# Use glob to get all the files ending with "_Summary_SEGID.csv"
file_paths = glob.glob(os.path.join(subfolder_path, "*.csv"))

# Create an empty list to store dataframes
dfs = []

# Loop through the files, parse filename, and read them into dataframes
for file_path in file_paths:
    # Get filename without extension
    filename = os.path.splitext(os.path.basename(file_path))[0]
    
    # Split filename based on double underscore
    modVersion, scnGroup, scnYear = filename.split('__')
    
    # Read the CSV into a dataframe
    df = pd.read_csv(file_path)
    df = df.loc[df['FUNCGROUP'] == 'Total']
    
    # Add new columns with values from filename
    df['modVersion'] = modVersion
    df['scnGroup'] = scnGroup
    df['scnYear'] = scnYear
    
    dfs.append(df)

# If you want to concatenate all the dataframes into a single dataframe
dfCombined = pd.concat(dfs, ignore_index=True)
dfCombined['scnYear'] = dfCombined['scnYear'].astype('int64')
dfCombined
```

```python
import json
import numpy as np

_df0 = dfCombined_melt_withkeys_pivot_atts.copy()

# Loop through scenarios in dfScn
for _, row in dfScn.iterrows():

    _scenarioCode = row['modVersion'] + '__' + row['scnGroup'] + '__' + str(row['scnYear'])
    print(_scenarioCode)

    # format scenario details for json export
    jsonScenarioDetails = row.to_dict()

    # Filter the main dataframe based on the scenario
    _df = _df0[(_df0['modVersion'] == row['modVersion']) & (_df0['scnGroup'] == row['scnGroup']) & (_df0['scnYear'] == row['scnYear'])].copy()
    _df.drop(columns=['modVersion','scnGroup','scnYear'], inplace=True)

    # GET LIST OF THE POSSIBLE UNIQUE FIELD COMBINATIONS OF DIMENSIONS, SO WHEREVER THERE IS A NON NULL OR EMPTY STRING VALUE

    # add list of unique values for each dCol
    _dfDimsWithOptions = dfDimensions.copy()
    _dfDimsWithOptions['dOptions'] = [_df[dCode].dropna().unique().tolist() for dCode in dfDimensions['dCode']]

    # get list of dimension groups (any column that begins with d)
    _lstDCols = [col for col in _df0.columns if col.startswith('d')]
    _dfDims = _df[_lstDCols].fillna('').drop_duplicates()

    # Get unique rows of combinations... a 1 means there are values in that column
    _dfDimsUniqCombos = _dfDims.applymap(lambda x: 0 if x=='' else 1).drop_duplicates()

    # Convert the unique combinations to a list of tuples
    _lstDimsUniqFieldCombos = list(_dfDimsUniqCombos.itertuples(index=False, name=None))

    # initialize
    _dfDimACodes = pd.DataFrame()
    jsonDataCombined = {}

    # loop through all unique field combinations in list
    for _dimCombo in _lstDimsUniqFieldCombos:

        # create string of column names that are part of given unique field combo separated by '_'
        _strFilterGroup = '_'.join([col for idx, col in enumerate(_lstDCols) if _dimCombo[idx] == 1])
        

        ## FILTER DATAFRAME BY ROWS THAT CORRESPOND TO UNIQUE FIELD COMBO

        # Define a function to convert to binary
        def is_empty_or_not(value):
            return 0 if value=='' else 1
        
        # Initialize the condition as True for all rows
        conditionMatchDimCombo = pd.Series([True] * len(_df), index=_df.index)

        # Apply the condition programmatically for each column in the list
        for col, comb_value in zip(_lstDCols, _dimCombo):
            conditionMatchDimCombo &= _df[col].apply(is_empty_or_not) == comb_value

        # Filter by condition and remove columns where all values are NaN -- I THINK THIS DROPS ATTRIBUTE COLUMNS THAT DON'T MATCH UNIQUE FILED COMBO CONDITION
        _filtered_df = _df[conditionMatchDimCombo].dropna(axis=1, how='all').copy()
        #display(_filtered_df)

        # create list of all non NA columns that remain so they can be dropped later... droping later as we get further into nesting of json
        _lstDColsToDrop = [col for col in _filtered_df.columns if col.startswith('d') and (_filtered_df[col] == '').all()]

        # Drop the identified columns - only have dCols with values
        _df2 = _filtered_df.drop(columns=_lstDColsToDrop)
        #display(_df2)

        # Get list of columns that begin with "a", append to dataframe of the dimension group for all possible combinations
        _dfDimComboACodes = pd.DataFrame([{"dimensions": _strFilterGroup, 'aCode': [col for col in _filtered_df.columns if col.startswith('a')]}])
        _dfDimComboACodes = _dfDimComboACodes.explode('aCode').reset_index(drop=True)
        #print(_dfDimComboACodes)

        # concat to 
        _dfDimACodes = pd.concat([_dfDimACodes, _dfDimComboACodes], ignore_index=True)


        # get unique filter values
        _lstDColsRemain = [col for col in _df2.columns if col.startswith('d')]
        #display(_lstDColsRemain)
        if _lstDColsRemain:
            _dfDimOptions = _df2[_lstDColsRemain].drop_duplicates()
        else:
            # add dummy row for no filter
            _dfDimOptions = pd.DataFrame({'nofilter': ['']})

        #display(_dfDimOptions)

        # Handle no filter data

        first_row = True

        for index, row in _dfDimOptions.iterrows():

            _filterRowIndex = '_'.join(row.astype(str).values)

            print(_filterRowIndex)
            
            final_condition = ""

            if _lstDColsRemain != []:
                # Dynamically construct the filter condition
                shared_columns = set(_dfDimOptions.columns) & set(_df2.columns)
                conditions = (_df2[col].isin([row[col]]) for col in shared_columns)
                final_condition = next(conditions)

                for condition in conditions:
                    final_condition &= condition

                filtered_df2 = _df2[final_condition].copy()
                filtered_df2.drop(columns=_lstDColsRemain, inplace=True)
            else: # no filter condition
                filtered_df2 =  _df2

            _df2noI = filtered_df2.reset_index()

            jsonDataForDimOption = {}

            for index, row in _df2noI.iterrows():
                # Initialize an empty dictionary for the row
                row_dict = {}

                # Loop through each column that starts with 'a'
                for col in _df2noI.columns[_df2noI.columns.str.startswith('a')]:
                    # Add to row_dict if value is greater than 0
                    if row[col] > 0:
                        row_dict[col] = row[col]
                
                # If row_dict is not empty, add it to jsonDataCombined with SEGID as the key
                if row_dict:
                    jsonDataForDimOption[row['SEGID']] = row_dict 
            
            # Encapsulate jsonDataCombined under the _filterRowIndex
            final_json = {_filterRowIndex: jsonDataForDimOption}

            ## Convert to a JSON formatted string
            #json_output = json.dumps(final_json)

            # Append the JSON structure from this iteration to the combined
            jsonDataCombined = {**jsonDataCombined, **final_json}

    _dfAttributesWFilter = pd.DataFrame.merge(dfAttributes, _dfDimACodes, on='aCode')
    jsonAttributes = _dfAttributesWFilter.to_dict(orient='records')
    jsonFilters = _dfDimsWithOptions.to_dict(orient='records')

    # Construct final JSON structure
    jsonScenario = {
        'scenarioDetails': jsonScenarioDetails,
        'attributes': jsonAttributes,
        'filters': jsonFilters,
        'data': jsonDataCombined#,
        #'aggregations': aggregations_list
    }

    # Specify the desired file name
    filename = '_site/data/scnData/' + _scenarioCode + '/roadway-trends.json'

    # Open the file in write mode
    with open(filename, 'w') as file:
        json.dump(jsonScenario, file)

```
