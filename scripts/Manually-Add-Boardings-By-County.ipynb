```python
import pandas as pd
from dbfread import DBF
```

```python
dfData = pd.DataFrame([
    ['RTP32v900'    , r"E:\GitHub\WF-TDM-v9x\Scenarios\v9.0 - official\RTP_2032\4_ModeChoice\3_TransitAssign\_v9_SE32_Net32_2_OD_Station.dbf"],
    ['RTP_2032_SLC', r"T:\WF-TDM-v9x\Scenarios\v9_SLC_TransitAnalysis\RTP_2032_SLC\4_ModeChoice\3_TransitAssign\_v9_SE32_HWY32_SLCoTransit50_2_OD_Station.dbf"],
    ['RTP_2032_SLC_Select'  , r"T:\WF-TDM-v9x\Scenarios\v9_SLC_TransitAnalysis\RTP_2032_SLC_Select\4_ModeChoice\3_TransitAssign\_v9_SE32_HWY32_SLCoTr50Sct_2_OD_Station.dbf"  ],
    ['RTP_2032_SLCo_Phase3' , r"T:\WF-TDM-v9x\Scenarios\v9_SLC_TransitAnalysis\RTP_2032_SLCo_Phase3\4_ModeChoice\3_TransitAssign\_v9_SE32_HWY32_SLCoTr50SLCo_2_OD_Station.dbf"],
], columns=['scenario','filename'])
dfData
```

```python
# List to hold the data from each file
all_data = []

# Iterate over the DataFrame
for _, row in dfData.iterrows():

    # Load the DBF file
    table = DBF(row['filename'], load=True)  # Set load=True to load all records into memory
    
    # Convert to DataFrame
    temp_df = pd.DataFrame(iter(table))

    # Add the 'Scenario' column to the DataFrame
    temp_df['scenario'] = row['scenario']
    
    # Append to the list
    all_data.append(temp_df)

# Concatenate all the DataFrames in the list
final_df = pd.concat(all_data, ignore_index=True)

# Display the resulting DataFrame
display(final_df)
```

```python

# Path to the DBF file
dbf_path = r"T:\WF-TDM-v9x\Scenarios\v9_SLC_TransitAnalysis\RTP_2032_SLCo_Phase3\0_InputProcessing\UpdatedMasterNet\GIS\MasterNet_v9 - 2023-09-16 - Node.dbf"

# Load the DBF file
table = DBF(dbf_path, load=True)  # Set load=True to load all records into memory

# Convert to DataFrame
df_nodes = pd.DataFrame(iter(table))

# Display the DataFrame
display(df_nodes)
```

```python
df_data_on_nodes = pd.merge(df_nodes, final_df, left_on='N', right_on='A')
df_data_on_nodes
```

```python
df_data_on_nodes_with_taz = pd.merge(df_data_on_nodes, df_taz, on='TAZID')
df_data_on_nodes_with_taz
```
